```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np


def how_many_days(year:int)->int:
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return 366
            else:
                return 365
        else:
            return 366
    else:
        return 365

# Astronomical year length in days
astronomical_year = 365.2425

def create_cumulative_days():
  years = [0] * 4001
  for i in range(1,4001):
    years[i] = years[i-1] + how_many_days(i)

  def cumulative_days(year):
    return years[year]

  return cumulative_days


cumulative_days = create_cumulative_days()
```

```python


# Create figure and axis
dataStart = int(input("Please inform first year (enter for 1730)>") or 1730 )
dataEnd = int(input("Please inform last year(enter for 2330)>") or 2330)
print(f"Starting in {dataStart} and ending in {dataEnd}")

# Generate years from dataStart, dataEnd+1
yearRange =  range(dataStart,dataEnd+1)
variations = [ (i * astronomical_year - cumulative_days(i)) for i in yearRange]

fig, ax = plt.subplots()
ax.set_xlim(dataStart, dataEnd+1)
ax.set_ylim(-2, +2)
ax.set_xlabel('Year')
ax.set_ylabel('Cumulative Difference (days)')
ax.set_title('Astronomical Year vs Calendar Year')

ax.plot( yearRange, variations )
ax.set_axis_on()
ax.set_yticks(np.arange(-2, 2, 0.5))
ax.axhline(y=0, color='k')
fig.set_size_inches(20, 5)
plt.show()

```
